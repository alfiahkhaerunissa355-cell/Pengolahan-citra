```python
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt
```

## 1. Akuisisi dan Representasi Citra

```python
# Membaca 3 Citra
img1 = cv2.imread('Pria.jpg')
img2 = cv2.imread('Patung.jpg')
img3 = cv2.imread('Rumah.jpg')

# Cek apakah semua gambar berhasil dimuat
if img1 is None or img2 is None or img3 is None:
    print("Error: Salah satu gambar tidak ditemukan.")
else:
    # Konversi BGR ke RGB
    img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
    img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
    img3 = cv2.cvtColor(img3, cv2.COLOR_BGR2RGB)

    # plt.figure agar gambar yag tampil tidak terlalu kecil
    plt.figure(figsize=(12,4))

    # plt.subplot untuk mengatur posisi gambar (1 kolom,3 baris,n = posisi)
    # plt.axis agar urutan gambar berbaris sejajar sesuai ukuran
    plt.subplot(1,3,1)
    plt.imshow(img1)
    plt.title("Pria")
    plt.axis('off')

    plt.subplot(1,3,2)
    plt.imshow(img2)
    plt.title("Patung")
    plt.axis('off')

    plt.subplot(1,3,3)
    plt.imshow(img3)
    plt.title("Rumah")
    plt.axis('off')

    plt.show()

    # Menampilkan ukuran citra
    print("Ukuran Pria:", img1.shape)
    print("Ukuran Patung:", img2.shape)
    print("Ukuran Rumah:", img3.shape)
```

```python

# Menampilkan jumlah channel tiap gambar
def get_channels(img):
    if len(img.shape) == 3:
        return img.shape[2]
    else:
        return 1

print("Jumlah channel Pria :", get_channels(img1))
print("Jumlah channel Patung :", get_channels(img2))
print("Jumlah channel Rumah  :", get_channels(img3))
print(" ")
print("Kalau grayscale → tidak punya channel ke-3, jadi dicek dengan len(img.shape) dan hasilnya akan : 1")
```

```python
# Tipe data Pixel
print("===== Data Gambar Pria =====")
print("Shape citra :", img1.shape)
print("Tipe data   :", img1.dtype)

print("===== Data Gambar Patung =====")
print("Shape citra :", img2.shape)
print("Tipe data   :", img2.dtype)

print("===== Data Gambar Rumah =====")
print("Shape citra :", img3.shape)
print("Tipe data   :", img3.dtype)
```

```python
# Representasi Citra dalam Matriks
print("Sebagian matriks citra RGB (10x10 pixel pertama):")
print(img1[0:10, 0:10])
```

# 2. Analisis Format Citra

```python
# Membaca 3 Citra
imgjpg = cv2.imread('Rumah.jpg')
imgpng = cv2.imread('Rumah.png')

# Cek apakah semua gambar berhasil dimuat
if imgjpg is None or imgpng is None:
    print("Error: Salah satu gambar tidak ditemukan.")
else:
    # Konversi BGR ke RGB
    imgjpg = cv2.cvtColor(imgjpg, cv2.COLOR_BGR2RGB)
    imgpng = cv2.cvtColor(imgpng, cv2.COLOR_BGR2RGB)

    # plt.figure agar gambar yag tampil tidak terlalu kecil
    plt.figure(figsize=(12,4))

    plt.subplot(1,2,1)
    plt.imshow(imgpng)
    plt.title("PNG")
    plt.axis('off')


    plt.subplot(1,2,2)
    plt.imshow(imgjpg)
    plt.title("JPG")
    plt.axis('off')

    plt.show()

    # Menampilkan Ukuran Citra
    size_png = os.path.getsize('Rumah.png') / 1024
    size_jpg = os.path.getsize('Rumah.jpg') / 1024

    print("=== Ukuran File ===")
    print(f"PNG : {size_png:.2f} KB")
    print(f"JPG : {size_jpg:.2f} KB")

    # Jenis Kompresi
    print("\n=== Jenis Kompresi ===")
    print("PNG : Lossless")
    print("JPG : Lossy")

    # Kualitas Visual
    imgjpg_resize = cv2.resize(imgjpg, (imgpng.shape[1], imgpng.shape[0]))

    mse = np.mean((imgpng.astype("float") - imgjpg_resize.astype("float")) ** 2)

    print("\n=== Kualitas Visual ===")
    print("MSE (semakin kecil semakin mirip):", mse)
```

# 3. Pengolahan Citra Berwarna

```python
# membaca citra
img_pria = cv2.imread('Pria.jpg')

# konversi BGR ke RGB (OpenCV default BGR)
img_rgb = cv2.cvtColor(img_pria, cv2.COLOR_BGR2RGB)

# Pemisahan Channel RGB
R = img_rgb[:,:,0]
G = img_rgb[:,:,1]
B = img_rgb[:,:,2]

plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.imshow(R, cmap='Reds')
plt.title("Red Channel")
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(G, cmap='Greens')
plt.title("Green Channel")
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(B, cmap='Blues')
plt.title("Blue Channel")
plt.axis('off')

plt.show()
```

```python
# Konversi RGB ke Grayscale
gray = cv2.cvtColor(img_pria, cv2.COLOR_BGR2GRAY)

plt.imshow(gray, cmap='gray')
plt.title("Citra Grayscale")
plt.axis('off')
plt.show()
```

```python
# Transformasi Warna
color_map = cv2.applyColorMap(gray, cv2.COLORMAP_JET)

plt.subplot(1,2,1)
plt.imshow(gray, cmap='gray')
plt.title("Grayscale")
plt.subplot(1,2,2)
plt.imshow(color_map)
plt.title("Pseudo Color")
plt.show()
```

# 4. Segmentasi Citra dan Pemisahan Background

```python

img_patung = cv2.imread("Patung.jpg")


# Masking
hsv = cv2.cvtColor(img_patung, cv2.COLOR_BGR2HSV)

# Semakin tinggi setiap nilai tiap kolom, Bagian Batang dan Pot Semakin hilang
lower = np.array([0,10,20])

# Kolom 1 : Semakin tinggi nilai, semakin terlihat bagian Pot nya
# Kolom 2 dan 3 : Semakin Rendah nilai, semakin gelap bagian Daun, Semakin tinggi nilai semakin banyak Noise BG yang tertangkap
upper = np.array([255,180,170])

mask = cv2.inRange(hsv, lower, upper)

result = cv2.bitwise_and(img_patung,img_patung,mask=mask)

# Convert hasil mask dan mask ke RGB
img_mask = cv2.cvtColor(img_patung, cv2.COLOR_BGR2RGB)
result_mask = cv2.cvtColor(result, cv2.COLOR_BGR2RGB)

# Tampilkan
plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.imshow(img_mask)
plt.title("Original")
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(mask, cmap='gray')
plt.title("Mask")
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(result_mask)
plt.title("Segmented")
plt.axis('off')

plt.show()
```

# 5. Deteksi Kontur

```python
# Dengan Metode Canny Edge Contour / Edge Detection

img_pesawat = cv2.imread('Pesawat.jpg')

img_copy = img_pesawat.copy()

# Cek apakah semua gambar berhasil dimuat
if img_pesawat is None:
    print("Error: Salah satu gambar tidak ditemukan.")
else:

  # Konversi RGB ke Grayscale ()
  gray_pesawat = cv2.cvtColor(img_pesawat, cv2.COLOR_BGR2GRAY)

  #Threshold
  _, binary_image = cv2.threshold (gray_pesawat, 245, 255, cv2.THRESH_BINARY)

  # Invert binary image
  invert_image = ~binary_image

  # Temukan kontur
  contours, hierarchy = cv2.findContours(invert_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

  # Gambar kontur pada gambar asli
  image_with_contours = cv2.drawContours(img_pesawat, contours, -1, (0, 0, 255), 3)

  titles = ['Original','Grayscale','Binary','Invert','Hasil Kontur']
  images = [img_copy,gray_pesawat,binary_image,invert_image,image_with_contours]

  # Tampilkan dengan matplotlib
  plt.figure(figsize=(12, 8))
  for i in range(5):
        plt.subplot(2, 3, i + 1)
        plt.imshow(images[i], cmap='gray')
        plt.title(titles[i])
        plt.axis('off')

  plt.tight_layout()
  plt.show()





```
